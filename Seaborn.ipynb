# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})

# sns.lineplot(x="var", y="var_1", data=x_1)
sns.lineplot(x=var,y=var_1)
plt.title("Using Seaborn")

plt.show()
```

```python
plt.plot(var,var_1)
plt.title("Using Matplotlib")
plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.set(style="darkgrid")
order_1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, hue_order=["Female","Male"],
            errorbar=("ci",90), n_boot=3, palette="icefire", saturation=0.7, err_kws={"color":"g","linewidth":5}, dodge=True, alpha=0.9)
plt.legend(loc=2)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="h")
plt.show()
```

# Violin Plot

### A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data accross several levels of  one (or more) categorical variables such that those distributions can be compared.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var 
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2, hue="day", palette="Dark2", saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, hue="time", order=["Dinner", "Lunch"], inner="quartile")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True, density_norm="count")
plt.show()
```

```python
sns.violinplot(x="total_bill", y="day", data=var, inner="stick")
```

```python
sns.violinplot(y="total_bill", data=var)
plt.show()
```

# Strip Plot

### Strip Plot is basically a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day", y="total_bill", data=var,hue="sex", linewidth=1, edgecolor="g", size=6, alpha=0.7)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var,hue="sex", jitter=0.3)
plt.show()
```

```python
sns.stripplot(y=var["total_bill"])
plt.show()
```

# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
o = ["Fri", "Sun", "Thur", "Sat"]
mp = {"marker":"+", "markeredgecolor":"w"}
sns.boxplot(x="day", y="total_bill", data=var, hue="time", order=o, showmeans=True, meanprops=mp, linewidth=3, palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
mp = {"marker":"+", "markeredgecolor":"w"}
sns.boxplot(data=var, showmeans=True, meanprops=mp, linewidth=3, palette="plasma", orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y = var["total_bill"])
plt.show()
```

# Facet Grid

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var, col="day", hue="sex")
fg.map(plt.scatter,"total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col="sex", hue="day")
fg.map(plt.bar,"total_bill", "tip").add_legend()
plt.show()
```
